```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Product Data
products = pd.DataFrame({
    "Product": [
        "Running Shoes",
        "Yoga Mat",
        "Dumbbells",
        "Gaming Mouse"
    ],
    "Features": [
        "Sports Fitness",
        "Fitness Yoga",
        "Fitness Gym",
        "Gaming Electronics"
    ]
})
```

```python
# Convert Features to Numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(products["Features"])
```

```python
# Similarity Matrix
similarity = cosine_similarity(X)
```

```python
# Customer purchased Yoga Mat
product_index = 1

scores = list(enumerate(similarity[product_index]))

recommendations = sorted(
    scores,
    key=lambda x: x[1],
    reverse=True
)[1:3]

print("Recommendations:")

for i, score in recommendations:
    print(products.iloc[i]["Product"])
```

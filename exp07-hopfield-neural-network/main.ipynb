```python
import matplotlib.pyplot as plt
```

```python
income = 60000

budget = {
    "Rent": 18000,
    "Groceries": 10000,
    "Transport": 5000,
    "Health": 5000,
    "Entertainment": 3000,
    "Savings": 12000,
    "Misc": 7000
}
```

```python
# Hopfield Energy Function
energy = 0

if budget["Savings"] < 12000:
    energy += 100

if budget["Rent"] > 18000:
    energy += 100

if budget["Health"] < 5000:
    energy += 100

if budget["Entertainment"] < 3000:
    energy += 100

print("Energy:", energy)
```

```python
# Recommendation
if energy == 0:
    print("Optimized Budget Found")
else:
    print("Adjust Budget")
```

```python
# What-if Analysis
print("\nWhat-if Analysis")
budget["Entertainment"] += 1000
budget["Savings"] -= 1000

print("New Entertainment:", budget["Entertainment"])
print("New Savings:", budget["Savings"])
```

```python
# Sensitivity Analysis
print("\nSensitivity Analysis")

for inc in [40000, 50000, 60000, 70000]:
    print(
        f"Income={inc}, Savings Goal={0.2*inc}"
    )
```

```python
# Visualization
plt.bar(
    budget.keys(),
    budget.values()
)

plt.title("Budget Allocation")
plt.xticks(rotation=45)
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Scrape WHO website
soup = BeautifulSoup(
    requests.get("https://www.who.int/news-room").text,
    "html.parser"
)
```

```python
# Web Content Mining
articles = [h.get_text(strip=True) for h in soup.find_all("h2")]
print("Articles:", articles[:5])
```

```python
# Web Structure Mining
links = soup.find_all("a", href=True)
print("Total Links:", len(links))
```

```python
# Web Usage Mining
pages = [a.get_text(strip=True) for a in links if a.get_text(strip=True)]
counts = pd.Series(pages).value_counts().head(5)

print("\nTop Pages:")
print(counts)
```

```python
# Visualization
counts.plot(kind="bar")
plt.show()
```

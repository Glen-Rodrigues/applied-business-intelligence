```python
import pandas as pd
from sqlalchemy import create_engine
from sklearn.linear_model import LinearRegression
```

```python
# SQLAlchemy Engine
engine = create_engine(
    "mysql+pymysql://root:root@localhost/retail_dss"
)
```

```python
# Read Data
query = "SELECT * FROM purchases"
df = pd.read_sql(query, engine)
```

```python
# Descriptive Analytics
df["total_amount"] = df["quantity"] * df["price"]

print("\nTotal Sales:")
print(df["total_amount"].sum())

print("\nSales by Category:")
print(df.groupby("category")["total_amount"].sum())
```

```python
# Predictive Analytics
monthly_sales = df.groupby(df["purchase_date"].astype(str))["total_amount"].sum()

X = [[i] for i in range(len(monthly_sales))]
y = monthly_sales.values

model = LinearRegression()
model.fit(X, y)

prediction = model.predict([[len(monthly_sales)]])

print("\nPredicted Future Sales:")
print(round(prediction[0], 2))
```

```python
# Prescriptive Analytics
top_category = df.groupby("category")["total_amount"].sum().idxmax()

print("\nRecommendation:")
print(f"Increase stock in {top_category}")
```
